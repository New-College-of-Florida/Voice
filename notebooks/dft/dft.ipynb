Fourier Transforms
============================

DFS &amp; DFT
-------------

The discrete Fourier transform (DFT) of a signal is defined using the following formula:

$$X(k)=\sum_{n=0}^{N-1} x(n) e^{-j(2\pi/N)(n)(k)},
k=0,\ldots,N-1.$$

A similar formula is used for the discrete Fourier series (DFS) of a periodic signal.

The DFS / DFT is a vector of complex numbers which can be
represented via their magnitude $|X(n)|$ (absolute value) and
phase $\arctan \frac{\text{Im}(X(n))}{\text{Re}(X(n))}$,
$n=0,\ldots,N-1$.

### Transformation matrix

DFS / DFT can be easily implemented using the transform matrix $\mathbb{W}_N$. We recall that for a signal of length $N$ the transform matrix has as $n$-th row and $k$-th column element $\mathbb{W}_N(n,k)=e^{-j(2\pi/N)(n)(k)}=W_N^{(n)(k)}$, with $n,k=0,\ldots,N-1$, and there we denote $W_N=e^{-j(2\pi/N)}$.

In Python, a DFT matrix can be generated using the following code.

```python
def dftmatrix(N):
    '''construct DFT matrix'''
    import numpy as np
    
    # create a 1xN matrix containing indices 0 to N-1
    a = np.expand_dims(np.arange(N), 0)
    
    # take advantage of numpy broadcasting to create the matrix
    WN = np.exp(-2j*np.pi*a.T*a/N)
    
    return WN
```

The transformation $X$ of the signal $x$, i.e. $X(k)$; is then given by (here using a short constant signal), multiplying the (here) $3\times 3$ DFT Matrix and the input signal (in its vector form).

```python
%pylab inline

N = 6
WN = dftmatrix(N)

x = np.ones(N)
X = np.dot(WN, x)
```

The inverse transformation $x(n)=\frac{1}{N} \sum_{k=1}^{N} X(k) e^{j(2\pi/N)(n)(k)}$ is then computed using:

```python
b = np.expand_dims(np.arange(N),0)
IN = (1.0/N)*np.exp(2j*np.pi*b.T*b/N)

x2 = np.dot(X,IN)
print 'Error: ',(np.abs(x - x2).sum())
print '(This small error is due to machine precision.)'
```

### Magnitude and Phase

Let's start with a simple signal: unitary step function (periodic)
of length N=128 (we shall see later what motivates us in choosing a
power of 2).

```python
x = np.append(np.ones(64),np.zeros(64))
```

We can see what this signal looks like in the next figure

```python
y = np.linspace(0,127,128)
# The plotting is done by the pylab.stem function
stem(y,x)
xlabel('Samples')
ylabel('x')
ylim([0,3])
show()
```

The transformation matrix and the transform is then

```python
x = x.flatten();
N = 128
W128 = dftmatrix(N)
X  = np.dot(W128,x)
```

 The stem plot of the magnitude (absolute value) of the DSF or DFT

```python
magnitude = abs(X)
```

is given by

```python
y = np.linspace(0,127,128)
y[:] = y[:]/len(y) #we are normalizing the frequencies
# Plotting is done by the pylab.stem function
stem(y,magnitude)
ylabel('Magnitude')
xlabel('Normalized Frequencies')
show()
```

### The Role of the Fourier transform 

The Fourier transformation has a twofold
reason to exist: (a) it lets us move into a domain where certain
computations are easier to implement (e.g., convolution) and (b) it
lets us represent the energy of the signal in the frequency
domain rather than the time domain. The second reason is
illustrated with a simple numerical example.

Load the file `frequencyRepresentation.mat`.
The signal $x$ contained in the file is a sum of sinusoids, that
is, is a sound composed of different tones. We would like to
understand how many tones or sinusoids compose the signal.

To import a `.mat` file in Python, use the following commands

```python
import scipy.io
mat = scipy.io.loadmat('frequencyRepresentation.mat')
print mat
```

And then,

```python
signal = mat['x']
print signal 
type(signal)
```

In the time domain we can either plot the signal (the figure below only displays 500 samples out of the 4000),

```python
signal = signal.reshape(signal.size,)
plot(np.arange(signal.size),signal)
xlim([0,500])
xlabel('Samples')
ylabel('x')
show()
```

or play it using the `scipy.io.wavfile.write` function.

But neither approach tells us how many sinusoids compose the signal.

Let's now take a look at the DFS /DFT. Create the transformation matrix (here $N=4000$) `W4000`, by

```python
signal = signal.reshape(signal.size,1)
W4000 = dftmatrix(4000)
X = np.dot(W4000,signal)
```

The stem plot in normalized frequencies yields:

```python
normFrequ = np.arange(1,X.size+1,dtype=float)/float(X.size)
plot(normFrequ,abs(X))
ylabel('|X|')
xlabel('Normalized Frequencies')
show()
```

It's made of two frequencies.

### DFS and DFT as projections on a basis of complex sinusoids

The DFS or DFT projects a signal of
length N on a basis of N complex sinusoids $W_N^{n,k}$ with
$k=1,\ldots,N$.

Let's see what happens if we reduce the dimension of the basis, that is the number of sinusoids used to represent the signal.

One complex sinusoid (of order $0$, i.e. a constant), taken from the first 128 samples of the unitary step function signal yields:

```python
x = np.append(np.ones(64),np.zeros(64))
W128 = dftmatrix(128)
X1 = np.dot(W128[0],x)
xx1 = np.real(np.conjugate(W128[:,0])*X1*1/128)
```

Now project onto the first 5, 25, and 100 complex sinusoids in the basis.

```python
X5 =  np.dot(W128[0:5],x) #analysis formula
xx5 = sum(np.real(np.conjugate(W128[:,0:5])*X5*1/128),axis=1) #Synthesis formula - that's why you need conjugate
#axis=1 so you sum the sinusoids and NOT just turn everything into one big scalar, i.e. "sum the second dimension", yet still have a
#128-sample signal rather than 5,25,100 128-sample signals

X25 = np.dot(W128[0:25],x)
xx25 = sum(np.real(np.conjugate(W128[:,0:25])*X25*1/128),axis=1)

X100 = np.dot(W128[0:100],x)
xx100 = sum(np.real(np.conjugate(W128[:,0:100])*X100*1/128),axis=1)
```

and plot the signal representation using 1, 5, 25, and 100 sinusoids, overlaid:

```python
plot(np.arange(xx1.size),xx1) #blue line - xx1 is actually a straight line through 0.5
plot(np.arange(xx5.size),xx5) #dark red line
plot(np.arange(xx25.size),xx25) #red line, more wiggly than X5
plot(np.arange(xx100.size),xx100) #periwinkle/cyan line
# Fill in here.

xlabel('Samples')
ylabel('x')
xlim([0,128])
show()
```

### Three tones
    
Let us consider another example. We create a simple signal $x$
composed of 3 tones $x_1$, $x_2$, and $x_3$

```python
import numpy as np

from scipy import constants as c
y = np.linspace(0,999,1000)
x1 = np.sin((y*2*c.pi*40)/1000)
x2 = np.sin((y*2*c.pi*80)/1000)
x3 = np.sin((y*2*c.pi*160)/1000)
```

We can put together these signals to get a final signal composed of $N=5000$ samples, with silence in between the 3 different sine waves.

This signal can be played by using Audacity, or the winsoud module.

```python
import scipy.io.wavfile


x = np.append(x1,[np.zeros(1000),x2,np.zeros(1000),x3])
scipy.io.wavfile.write("test.wav",22500,x)
```

Let's generate the Fourier matrix $W_N$ (with $N=5000$) and
compute the DFS or DFT of the signal

```python
W5000 = dftmatrix(5000)
X = np.dot(W5000,x)
```

The magnitude of the Fourier transformation is

```python
normFrequ = np.arange(1,X.size+1,dtype=float)/float(X.size)
plot(normFrequ,abs(X))
ylabel('|X|')
xlabel('Normalized Frequencies')
show()
```

where we can clearly see the three tones.

Let's now reduce the basis over which the signal is represented, and take:

```python
W5000reduced = np.append(W5000[0:600],W5000[4399:5000],axis=0)
iW5000reduced = np.conj(np.append(W5000[:,0:600],W5000[:,4399:5000],axis=-1))

Xapprox = np.dot(W5000reduced,x) ## Fill in here ##  # DFT/DFS on a reduced base
xapprox = iW5000reduced*Xapprox*1/128 ## Fill in here ##  # inverse DFT/DFS of the DFT/DFS on a reduced base
xapprox = sum(np.real(xapprox),axis=1) # real value of the result to get rid of imaginary numerical errors
scipy.io.wavfile.write("test2.wav",22500,xapprox) #now one of the three frequencies is missing!
```

So `xapprox` is the signal $x$ projected on a reduced base

Play the signal `xapprox` and take a look at its DFS / DFT:

```python
X = np.dot(W5000,xapprox)

plot(np.arange(1,5001,dtype=float)/float(5000),abs(X))
show() #and the plot confirms it.
```

You will notice the difference!

### DFT And Truncating

Consider a signal x composed of N samples and let's use the function
X = np.fft.fft(x,M)
to compute the FFT.

If M<N then the FFT is computed using the first M samples of x.
Let's take another two tones signal, with N=4000,

```python
import scipy.constants as c
import scipy.io.wavfile

y = np.linspace(0,3999,4000)
x1 = np.sin((y*2*c.pi)/25) # Fill in sin of 2 pi n / 25 
x2 = np.sin((y*2*c.pi)/20) # Fill in sin of 2 pi n / 20
x = x1 + x2
scipy.io.wavfile.write("test3.wav",12500,x)
```

that we can play using the `scipy.io.wavfile.write` function.
 
Let's plot the spectrum and represent the frequency axis relative
to the number of samples, i.e., between 0 and 1. We are going to
compute the fft using all the samples, that is

```python
M = 4000
X = np.fft.fft(x,M)
#Plot versus the normalized frequencies
normFrequ = np.arange(1,X.size+1,dtype=float)/float(M)# Fill in here.
plot(normFrequ,abs(X))
ylabel('|X|')
xlabel('Normalized Frequencies')
show()
```

As expected, the spectrum shows the energy of the signal concentrated on two precise frequencies.

Let us now reduce the number of samples used to compute the DFS / DFT, taking $M=500$ and $M=50$ and overlaying the plots: 

```python
# Fill in here:
X500 = np.fft.fft(x,500) #only first 500 samples
X50 = np.fft.fft(x,50) #only first 50 samples

normX500 = np.arange(1,X500.size+1,dtype=float)/float(500)
normX50 = np.arange(1,X50.size+1,dtype=float)/float(50)
plot(normX500,abs(X500))
plot(normX50,abs(X50))
plot(normFrequ,abs(X))
ylabel('|X|')
xlabel('Normalized Frequencies')
ylim([0,35])
show()
```

### DFT And Zero Padding

When $M > N$ the number of samples of the signal is increased to $M$ by adding zeros.

Since truncating decreases the resolution of the DFS / DFT, one might think that zero padding increases such a resolution. But we shall see that this is not the case.

So let's take the DFS / DFT computed using only 50 samples and reconstruct the signal using such a DFS / DFT.

```python
x50 = np.fft.ifft(X50)
```

where we have used here the inverse FFT function `scipy.fftpack.ifft()`.
 
We have obtained a signal $x_{50}$ of $N=50$ samples and we compute the FFT with $M=4000$, that is, we zero pad the signal so as to have the initial number of samples before computing the FFT.

```python
M = 4000
X = np.fft.fft(x50,M)

#Plot versus the normalized Frequencies
normFrequ = np.arange(1,M+1,dtype=float)/float(M)
plot(normFrequ,abs(X))
ylabel('|X|')
xlabel('Normalized Frequencies')
show()
```

As shown in the plot, the two spectral lines are still not distinguishable even if the FFT has now been computed using 4000 samples.
